```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Pandas:",pd.__version__)
print("Numpy:",pd.__version__)
print("Matplotlib:",pd.__version__)
```

```python
df=pd.read_csv("netflix_titles.csv")
df=pd.read_csv("netflix_titles.csv",index_col=0)
df.head()

```

```python
df.shape
df.info()
df.isnull().sum()
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['director']=df["director"].fillna('Unknown')
df["cast"]=df["cast"].fillna('Unknown')
df['country']=df["country"].fillna('Unknown')
df['rating']=df["rating"].fillna('Unknown')
df['duration']=df["duration"].fillna('Unknown')
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
df.isnull().sum()
```

```python
df[df['date_added'].isnull()][['title','date_added']].head(10)
```

```python
df=df.dropna(subset="date_added")
df.isnull().sum()
```

```python
df["year_added"]=df["date_added"].dt.year
df["month_added"]=df["date_added"].dt.month   #.dt access date time properties
df.head()
```

```python
df[["date_added","month_added","year_added"]].head()
```

```python
df.head(-10)
```

```python
counts=df["type"].value_counts()
print(counts)
ratio=counts["Movie"]/counts["TV Show"]
np.round(ratio,1)

```

It appears in fact that netflix has more movies than series. The number is almost 2.4 times the number of series.

```python
movie_trends=df[df["type"]=="Movie"]["year_added"].value_counts().sort_index()
series_trends=df[df["type"]=="TV Show"]["year_added"].value_counts().sort_index()


yearly_total=df["year_added"].value_counts().sort_index()
yearly_total

import matplotlib.pyplot as plt

plt.figure()
plt.plot(yearly_total.index,yearly_total.values)
plt.title("Total Netflix Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of titles being added")
plt.show()
```

There's no abrupt increase in the rise of movies from 2008 - 2014. The increase in the rate of content being added started from 2015 and rose high upto 2019 (around 125 movies to 2000) 2000 movies were added in 2019 which was the peak and later it dropped but still higher than the starting years

```python
movie_trends=df[df["type"]=="Movie"]["year_added"].value_counts().sort_index()
series_trends=df[df["type"]=="TV Show"]["year_added"].value_counts().sort_index()

plt.figure()
plt.plot(movie_trends.index,movie_trends.values)
plt.plot(series_trends.index,series_trends.values)
plt.title("Movies Vs Series Trends")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.legend(["Movies","TV Shows"])
plt.show()
```

`*Graph Overview*`
Netflix started its production with movies.During the early years only movies were added .Around 2010 , series were added too because Netflix created an habit to their customers to watch and spend time in their website. So people were habituated with longer duration series too. There wasn't any abrupt increase in both of the graphs upto 2015. From 2015 we can see the increase in the graph meaning the movies and series are being added heavily upto 2019. But still the rate of movies being added is much higher than the rate of series being added. Movies peak was in 2019 , where around 2000 movies were added in a single year whereas the peak of Series was in 2020, where 600 series were added.
```python
df["listed_in"] = df["listed_in"].astype(str)
df["listed_in"] = df["listed_in"].str.split(", ")
df_exploded=df.explode("listed_in")
df_exploded["listed_in"].value_counts().head(10)

```

```python
df_exploded.groupby(["type", "listed_in"]).size().sort_values(ascending=False).head(10)
```

```python
top_genres=df_exploded["listed_in"].value_counts().head(10)
top_genres.plot(kind='pie')
plt.title("Top Netflix Genres")
plt.xlabel("Genres")
plt.ylabel("Count")
plt.show()
```

```python
top_genres=df_exploded["listed_in"].value_counts().head(10)
top_genres.plot(kind='bar')
plt.title("Top Netflix Genres")
plt.xlabel("Genres")
plt.ylabel("Count")
plt.show()
```

```python
df["country"]=df["country"].fillna("Unknown")
df["country"]=df["country"].str.split(",")
df_country=df.explode("country")
df_country["country"].value_counts().head(10)
```

```python
top_countries=df_country["country"].value_counts().head(10)
top_countries.plot(kind="bar")
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Countries")
plt.ylabel("Count")
plt.show()
```
